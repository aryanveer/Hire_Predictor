```python
from tkinter import *
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB
%gui tk
import pymysql
import pymysql.cursors
root=Tk()
root.geometry('900x600')
root.title("Hire Predictor")


label_0=Label(root, text="Hire Prediction Form",width=20,font=("bold",20), fg='BLUE')
label_0.place(x=300,y=30)

label_1=Label(root, text="NAME",width=20,font=("bold",10))
label_1.place(x=8,y=100)
a = StringVar()
e1 = Entry(root, textvariable = a)
e1.place(x=150,y=100)

label_2=Label(root, text="PERCENTAGE",width=20,font=("bold",10))
label_2.place(x=8,y=140)
b = IntVar()
e2=Entry(root, textvariable = b)
e2.place(x=150,y=140)

label_3=Label(root, text="BACKLOGS",width=20,font=("bold",10))
label_3.place(x=8,y=180)
c = IntVar()
e3 = Entry(root, textvariable = c)
e3.place(x=150,y=180)

label_4=Label(root, text="INTERNSHIP",width=20,font=("bold",10))
label_4.place(x=8,y=220)
d = IntVar()
e4=Entry(root, textvariable = d)
e4.place(x=150,y=220)

label_5=Label(root, text="FIRST ROUND",width=20,font=("bold",10))
label_5.place(x=8,y=260)
e = IntVar()
e5=Entry(root, textvariable=e)
e5.place(x=150,y=260)

label_6=Label(root, text="COMMUNICATION",width=20,font=("bold",10))
label_6.place(x=8,y=300)
f = IntVar()
e6=Entry(root, textvariable = f)
e6.place(x=150,y=300)

def take_values():
    candidate_info = []
    #num1=int(b.get())
    #num3=int(c.get())
    #num3=int(d.get())
    #num4=int(e.get())
    #num5=int(f.get())
    candidate_info.append(b.get())
    candidate_info.append(c.get())
    candidate_info.append(d.get())
    candidate_info.append(e.get())
    candidate_info.append(f.get())

    np_candidate_values = np.array([candidate_info])
    return np_candidate_values

final_list = []

#---------------------------------------------------------------------------------------------------------

# Machine Learning

df = pd.read_excel('Hired_Data_Set.xlsx')

dic = {'Yes':1,'No':0}
df['Hire?'] = df['Hire?'].map(dic)

col_names = ['Percentage','Backlog','Internship','First Round','Communication Skills','Hire?']
data = df[col_names]

#data to train the models
X_columns = ['Percentage','Backlog','Internship','First Round','Communication Skills']
X = data[X_columns]
y = data['Hire?']

model = DecisionTreeClassifier(random_state = 0)
model.fit(X, y)

forest = RandomForestClassifier(n_estimators=10, random_state=2)
forest.fit(X, y)
    
svc_2 = SVC(kernel="linear",C=10)
svc_2.fit(X,y)

logit = LogisticRegression(C=7).fit(X, y)

kmeans = KMeans(n_clusters = 2)
kmeans = kmeans.fit(X)

nb_classifier = GaussianNB()
nb_classifier.fit(X,y)
# Decision Tree Classifier
def decision_tree():
    candidate_values = take_values()
    #model = DecisionTreeClassifier(random_state = 0)
    #model.fit(X, y)
    predicted1 = model.predict(candidate_values)
    f1.set(predicted1[0])
    final_list.append(predicted1[0])

# Random Forest Classifier
def Random_Classifier():
    candidate_values = take_values()
    #forest = RandomForestClassifier(n_estimators=10, random_state=2)
    #forest.fit(X, y)
    predicted2 = forest.predict(candidate_values)
    f2.set(predicted2[0])
    final_list.append(predicted2[0])

# Logistic Regressor
def logistic():
    candidate_values = take_values()
    #logit = LogisticRegression(C=7).fit(X, y)
    predicted3 = logit.predict(candidate_values)
    f3.set(predicted3[0])
    final_list.append(predicted3[0])
    
# Support Vector Classifier
def svc():
    candidate_values = take_values()
    #svc_2 = SVC(kernel="linear",C=10)
    #svc_2.fit(X,y)
    predicted4 = svc_2.predict(candidate_values)
    f4.set(predicted4[0])
    final_list.append(predicted4[0])
    
# K-Means Clustering
def k_means():
    candidate_values = take_values()
    #kmeans = KMeans(n_clusters = 2)
    #kmeans = kmeans.fit(X)
    predicted5 = kmeans.predict(candidate_values)
    f5.set(predicted5[0])
    final_list.append(predicted5[0])
    
# Naive-Bayes Classifier
def naive_bayes():
    candidate_values = take_values()
    #nb_classifier = GaussianNB()
    #nb_classifier.fit(X,y)
    predicted6 = nb_classifier.predict(candidate_values)
    f6.set(predicted6[0])
    final_list.append(predicted6[0])
#

#---------------------------------------------------------------------------------------------------------

f1 = StringVar()
e7=Entry(root, textvariable = f1)
e7.place(x=10,y=360)
label_7=Label(root, text="Decision Tree",width=20,font=("bold",10))
label_7.place(x=10,y=380)

f2 = StringVar()
e8=Entry(root, textvariable = f2)
e8.place(x=160,y=360)
label_8=Label(root, text="Random Forest",width=20,font=("bold",10))
label_8.place(x=160,y=380)

f3 = StringVar()
e9=Entry(root, textvariable = f3)
e9.place(x=310,y=360)
label_9=Label(root, text="Logistic",width=20,font=("bold",10))
label_9.place(x=310,y=380)

f4 = StringVar()
e10=Entry(root, textvariable = f4)
e10.place(x=460,y=360)
label_10=Label(root, text="SVC",width=20,font=("bold",10))
label_10.place(x=460,y=380)

f5 = StringVar()
e11=Entry(root, textvariable = f5)
e11.place(x=610,y=360)
label_11=Label(root, text="K-Means",width=20,font=("bold",10))
label_11.place(x=610,y=380)

f6 = StringVar()
e12=Entry(root, textvariable = f6)
e12.place(x=760,y=360)
label_12=Label(root, text="Naive-Bayes",width=20,font=("bold",10))
label_12.place(x=760,y=380)


Button(root, text='Decision Tree',width=15,bg='black',fg='white',command=decision_tree).place(x=450, y=100)
Button(root, text='Random Forest',width=15,bg='BLACK',fg='white',command=Random_Classifier).place(x=450, y=140)
Button(root, text='Logistic Regressor',width=15,bg='BLACK',fg='white',command=logistic).place(x=450, y=180)
Button(root, text='SVC',width=15,bg='BLACK',fg='white',command=svc).place(x=450, y=220)
Button(root, text='K-Means',width=15,bg='BLACK',fg='white',command=k_means).place(x=450, y=260)
Button(root, text='Naive-Bayes',width=15,bg='BLACK',fg='white',command=naive_bayes).place(x=450, y=300)


def register():
    if(sum(final_list) >= 3):
        final.set('1')
        y_predicted = 1
    else:
        final.set('0')
        y_predicted = 0
    del final_list[:]
    
final = StringVar()
e12=Entry(root, textvariable = final)
e12.place(x=400,y=460)
label_12=Label(root, text="Result",width=20,font=("bold",10),justify='left')
label_12.place(x=380,y=490)


Button(root, text='Final Result',width=30,bg='BLUE',fg='white',command=register).place(x=350, y=530)

root.mainloop()
```
